```python
from sklearn.impute import SimpleImputer

simp = SimpleImputer(strategy='mean', missing_values=np.nan)
simp = simp.fit(data[['Other_Sales']])
data['Other_Sales'] = simp.transform(data[['Other_Sales']])

data.isnull().mean()*100
data.info()
```
