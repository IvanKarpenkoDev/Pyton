```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np
```

```python
data = pd.read_csv('gsales_6.csv')
data
```

```python
# Демонстрация графика количества значений категориального атрибута (countplot)
sns.countplot(x='Platform', data=data)
plt.title('Countplot of Attribute')
plt.show()
```

Строит график countplot, который показывает количество наблюдений для каждой категории атрибута "Platform" и добавляет заголовок "Countplot of Attribute" к графику.

```python
# Отображение среднего значения атрибутов (barplot)
sns.barplot(x='Other_Sales', y='Global_Sales', data=data)
plt.title('Mean Value of Attributes')
plt.show()
```

Строит график barplot, который показывает средние значения атрибутов "Other_Sales" и "Global_Sales" и добавляет заголовок "Mean Value of Attributes" к графику.

```python
# Реализация ящика с усами (boxplot)
sns.boxplot(x='Global_Sales', y='JP_Sales', data=data)
plt.title('Boxplot of Attributes')
plt.show()
```

использую библиотеку Seaborn для визуализации распределения значений двух числовых атрибутов с помощью графика ящика с усами (boxplot).

```python
# Создание тепловой карты корреляции (heatmap)
numeric_columns = data.select_dtypes(include=[np.number])
correlation = numeric_columns.corr()

sns.heatmap(correlation, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

код создает тепловую карту корреляции с помощью библиотеки Seaborn.

```python
# Показать соотношение между атрибутами (pairplot или pairgrid)
sns.pairplot(data)
plt.title('Pairplot of Attributes')
plt.show()
```

Код создает график парных взаимосвязей между всеми числовыми атрибутами с помощью библиотеки Seaborn.

```python
# Кластеризация с использованием KMeans
X = data[['NA_Sales', 'Other_Sales', 'Global_Sales']]  # Выберите нужные атрибуты для кластеризации
kmeans = KMeans(n_clusters=3,  n_init=10)  # Задайте количество кластеров
kmeans.fit(X)
```

```python
# Результаты кластеризации
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
data['cluster'] = labels
```

В коде производится кластеризация данных с использованием алгоритма K-средних (K-Means).

```python
# Описание результатов
cluster_counts = data['cluster'].value_counts()
print('Cluster Counts:')
print(cluster_counts)
```

Здесь производится подсчет количества элементов в каждом кластере после кластеризации.Вывожу в тоге на экран количество элементов в каждом кластере после выполнения кластеризации.
